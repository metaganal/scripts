```python
countrylist = Table.columns[1:].tolist()
```

```python
from collections import OrderedDict
country_code = list(OrderedDict.fromkeys([i[:2] for i in countrylist]))
```

```python
country_code
```

```python
Table_by_country = []

for i in country_code:
    Table_by_country.append(Table.filter(like=i, axis=1))
```

```python
Table_by_country
```

```python
Tablecountry = pd.concat(Table_by_country, axis=1, keys=country_code)
```

```python
genusseries = Table.iloc[:,0].tolist()
```

```python
Tablecountry.index = genusseries
```

```python
for i in range(len(Table_by_country)):
    Table_by_country[i] = pd.concat([Table.iloc[:,0], Table_by_country[i]], axis=1)
    Table_by_country[i].rename(columns={'Unnamed: 0': 'Genus'}, inplace=True)
```

```python
len(country_code)
```

```python
for i in range(len(Table_by_country)):
    filename = '/Users/felix/kadai2' + country_code[i] + '.txt'
    Table_by_country[i].to_csv(filename, sep='\t', mode='w+')
```

```python
import scipy
import scipy.stats
from scipy.stats import mannwhitneyu as u_test
```

```python
tabletuple = [(country_code[i], Table_by_country[i]) for i in range(len(country_code))]
```

```python
countrytotabledict = {k:v for k,v in tabletuple}
```

```python
tmpdf = countrytotabledict['JP']
JPsamplelist = tmpdf.loc[tmpdf['Genus'] == 'Bacteroides'].iloc[0, 1:].tolist()

tmpdf = countrytotabledict['IT']
ITsamplelist = tmpdf.loc[tmpdf['Genus'] == 'Bacteroides'].iloc[0, 1:].tolist()
```

```python
ITsamplelist
```

```python
u_test(JPsamplelist, ITsamplelist, alternative='two-sided')
```
